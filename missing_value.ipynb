```python
import pandas as pd
test=pd.read_csv('test.csv')
train=pd.read_csv('train.csv')
conv_dict={'HomePlanet':{'Europa':0,'Earth':1,'Mars':2},
       'CryoSleep':{False:0,True:1},
       'Destination':{'TRAPPIST-1e':0,'55 Cancri e':1,'PSO J318.5-22':2},
      'VIP':{True:1,False:0}

     }
test=test.replace(conv_dict)
test=test.drop(['Name'],axis=1)
train=train.replace(conv_dict)
train=train.drop(['Name'],axis=1)
```

```python
train[['num','deck','side']]=train['Cabin'].str.split('/',expand=True)
train[['id','group']]=train['PassengerId'].str.split('_',expand=True)
train=train.drop(['PassengerId','Cabin'],axis=1)
```

```python
train['deck']=train['deck'].astype(float)
train['id']=train['id'].astype(float)
train['group']=train['group'].astype(float)
train['side']=train['side'].replace({'P':0,'S':1})
train['num']=train['num'].replace({'F':0,'G':1,'E':2,'B':3,'C':4,'D':5,'A':6,'T':7})
```

```python
test[['num','deck','side']]=test['Cabin'].str.split('/',expand=True)
test[['id','group']]=test['PassengerId'].str.split('_',expand=True)
test=test.drop(['PassengerId','Cabin'],axis=1)
```

```python
test['deck']=test['deck'].astype(float)
test['id']=test['id'].astype(float)
test['group']=test['group'].astype(float)
test['side']=test['side'].replace({'P':0,'S':1})
test['num']=test['num'].replace({'F':0,'G':1,'E':2,'B':3,'C':4,'D':5,'A':6,'T':7})
```

```python
train_new = test[['CryoSleep','FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']]
X = train_new[train_new['CryoSleep'] == 1.0]
X = X[['FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']]
X = X.fillna(0)




train_new.loc[X.index, ['FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']] = X
test[['CryoSleep','FoodCourt','RoomService','Spa','ShoppingMall','VRDeck']]=train_new



train_new0=train_new[['CryoSleep']]
train_new0['Sum']=train_new[['FoodCourt','RoomService','Spa','ShoppingMall','VRDeck']].sum(axis=1)
y=train_new0[train_new0['Sum']==0.0]
y=y.fillna(1)
train_new.loc[y.index,['CryoSleep']]=y['CryoSleep']
z=train_new0[train_new0['Sum']!=0.0]
z=z.fillna(0.0)
train_new.loc[z.index,['CryoSleep']]=z['CryoSleep']
test[['CryoSleep','FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']]=train_new
```

```python
train_new = train[['CryoSleep','FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']]
X = train_new[train_new['CryoSleep'] == 1.0]
X = X[['FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']]
X = X.fillna(0)




train_new.loc[X.index, ['FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']] = X
train[['CryoSleep','FoodCourt','RoomService','Spa','ShoppingMall','VRDeck']]=train_new



train_new0=train_new[['CryoSleep']]
train_new0['Sum']=train_new[['FoodCourt','RoomService','Spa','ShoppingMall','VRDeck']].sum(axis=1)
y=train_new0[train_new0['Sum']==0.0]
y=y.fillna(1)
train_new.loc[y.index,['CryoSleep']]=y['CryoSleep']
z=train_new0[train_new0['Sum']!=0.0]
z=z.fillna(0.0)
train_new.loc[z.index,['CryoSleep']]=z['CryoSleep']
train[['CryoSleep','FoodCourt','RoomService','ShoppingMall','Spa','VRDeck']]=train_new
```

```python
from sklearn.ensemble import RandomForestClassifier
data=test.drop(['RoomService', 	'FoodCourt', 	'ShoppingMall', 	'Spa', 	'VRDeck','Age'],axis=1)
for name in data.columns:
  if data[name].isna().sum()==0:
    continue
  X=data.loc[data[name].dropna().index,:].drop([name],axis=1)
  X=X.apply(lambda col: col.fillna(col.mode()[0]),axis=0)
  y=data[name].dropna()
  model=RandomForestClassifier(n_estimators=40,max_depth=10,min_samples_split=5)
  model.fit(X.values,y.values)
  print(name)
  pred=model.predict(data[data[name].isna()].drop([name],axis=1).fillna(-5).values)
  X=data[name]
  X[X.isna()]=pred
  data[name]=X

test[['HomePlanet', 	'CryoSleep', 	'Destination', 	'VIP', 	'num', 	'deck', 	'side', 	'id', 	'group']]=data
```
