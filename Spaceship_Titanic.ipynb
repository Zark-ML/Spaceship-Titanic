```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
train=train.drop(['Name'],axis=1)
test=test.drop(['Name'],axis=1)


conv_dict={'HomePlanet':{'Europa':0,'Earth':1,'Mars':2},
       'CryoSleep':{False:0,True:1},
       'Destination':{'TRAPPIST-1e':0,'55 Cancri e':1,'PSO J318.5-22':2},
      'VIP':{True:1,False:0}

     }

train=train.replace(conv_dict)
test=test.replace(conv_dict)


train['Transported']=train['Transported'].replace({True:1,False:0})



```

```python
train[['num','deck','side']]=train['Cabin'].str.split('/',expand=True)
train[['id','group']]=train['PassengerId'].str.split('_',expand=True)
train=train.drop(['PassengerId','Cabin'],axis=1)
```

```python
train['deck']=train['deck'].astype(float)
train['id']=train['id'].astype(float)
train['group']=train['group'].astype(float)
train['side']=train['side'].replace({'P':0,'S':1})
train['num']=train['num'].replace({'F':0,'G':1,'E':2,'B':3,'C':4,'D':5,'A':6,'T':7})
```

```python
test[['num','deck','side']]=test['Cabin'].str.split('/',expand=True)
test[['id','group']]=test['PassengerId'].str.split('_',expand=True)
test=test.drop(['PassengerId','Cabin'],axis=1)
```

```python
test['deck']=test['deck'].astype(float)
test['id']=test['id'].astype(float)
test['group']=test['group'].astype(float)
test['side']=test['side'].replace({'P':0,'S':1})
test['num']=test['num'].replace({'F':0,'G':1,'E':2,'B':3,'C':4,'D':5,'A':6,'T':7})
```
